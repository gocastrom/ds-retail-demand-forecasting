# 1. Exploración de Datos (EDA)

El objetivo de este notebook es realizar un análisis exploratorio inicial sobre los datos de ventas. Queremos entender:

- La distribución de las ventas.
- Las tendencias a lo largo del tiempo.
- La estacionalidad (semanal, mensual).
- La relación con las promociones.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de visualización
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (15, 8)

print('Cargando datos procesados...')
# Cargar una muestra para no sobrecargar la memoria si el archivo es muy grande
df = pd.read_parquet('../data/processed/full_data.parquet')
print('Datos cargados.')
```

## Tendencia General de Ventas

```python
# Agrupar ventas por fecha
sales_by_date = df.groupby('date')['unit_sales'].sum().reset_index()

# Graficar
plt.figure(figsize=(20, 7))
sns.lineplot(x='date', y='unit_sales', data=sales_by_date)
plt.title('Ventas Totales a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Unidades Vendidas (Total)')
plt.show()
```
